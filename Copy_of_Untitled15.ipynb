```python
# ner_user_input.py

import spacy

def main():
    # Load the pre-trained English model
    try:
        nlp = spacy.load("en_core_web_sm")
    except OSError:
        print("Model not found. Downloading 'en_core_web_sm' model...")
        import subprocess
        subprocess.run(["python", "-m", "spacy", "download", "en_core_web_sm"])
        nlp = spacy.load("en_core_web_sm")

    while True:
        # Get user input
        text = input("\nEnter a sentence for NER (or type 'exit' to quit): ")

        if text.lower() == 'exit':
            print("Exiting NER program.")
            break

        # Process the input text
        doc = nlp(text)

        # Display named entities
        if doc.ents:
            print("\nNamed Entities:")
            for ent in doc.ents:
                print(f"{ent.text} ({ent.label_})")
        else:
            print("No named entities found.")

if __name__ == "__main__":
    main()
```

```python
# text_summarizer.py

from transformers import pipeline

def main():
    # Load the summarization pipeline
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

    print("Text Summarization Program (type 'exit' to quit)\n")

    while True:
        # Get user input
        text = input("Enter text to summarize:\n")

        if text.lower() == "exit":
            print("Exiting the summarizer.")
            break

        if len(text.split()) < 20:
            print("Please enter at least 20 words for a meaningful summary.")
            continue

        # Generate summary
        summary = summarizer(text, max_length=130, min_length=30, do_sample=False)

        print("\n🔍 Summary:")
        print(summary[0]['summary_text'])
        print()

if __name__ == "__main__":
    main()
```

```python
# ⚙️ Step 2: Set up a fast Q&A model (DistilBERT)
from transformers import pipeline

# Load question-answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
```

```python
# 🧠 Step 3: Define medical assistant function
def answer_question(context, question):
    if not context or not question:
        return "Please provide both context and a question."
    result = qa_pipeline(question=question, context=context)
    return f"Answer: {result['answer']} (Confidence: {result['score']:.2f})"
```

```python
# 🎛️ Step 4: Build interactive UI using Gradio
import gradio as gr

gr.Interface(
    fn=answer_question,
    inputs=[
        gr.Textbox(lines=6, label="Medical Context", placeholder="Describe your symptoms or paste medical text..."),
        gr.Textbox(lines=2, label="Your Question", placeholder="Ask something about the symptoms...")
    ],
    outputs="text",
    title="🩺 Fast Medical Q&A Assistant",
    description="Enter a medical description and ask a question. This uses a lightweight BERT model for Q&A."
).launch(share=True, debug=True)
```

```python
from pydantic import BaseModel, validator, ValidationError
from typing import Optional
from datetime import date

# Stage 1: Raw Data (Extracted)
class RawPatientData(BaseModel):
    patient_id: str
    age: int
    gender: str
    symptom_description: str
    admission_date: Optional[str]

    @validator("gender")
    def validate_gender(cls, v):
        if v.lower() not in ("male", "female", "other"):
            raise ValueError("Gender must be male, female, or other")
        return v.lower()

# Stage 2: Cleaned Data (after transformation)
class CleanedPatientData(BaseModel):
    patient_id: str
    age: int
    gender: str
    symptoms: list[str]
    admission_date: Optional[date]

    @validator("symptoms", pre=True)
    def split_symptoms(cls, v):
        if isinstance(v, str):
            return [s.strip().lower() for s in v.split(",")]
        return v

# Stage 3: Preprocessed Data (ready for ML)
class PreprocessedPatientData(BaseModel):
    patient_id: str
    age: float
    gender_encoded: int
    symptom_vector: list[int]
```

```python
from datetime import datetime

# Input raw data
raw = {
    "patient_id": "P123",
    "age": 45,
    "gender": "Male",
    "symptom_description": "Fever, headache, fatigue",
    "admission_date": "2024-06-15"
}

try:
    # Step 1: Validate Raw
    raw_data = RawPatientData(**raw)

    # Step 2: Clean and Validate
    cleaned_data = CleanedPatientData(
        patient_id=raw_data.patient_id,
        age=raw_data.age,
        gender=raw_data.gender,
        symptoms=raw_data.symptom_description,
        admission_date=datetime.strptime(raw_data.admission_date, "%Y-%m-%d").date()
    )

    # Dummy encoding logic
    gender_map = {"male": 0, "female": 1, "other": 2}
    symptom_vocab = ["fever", "headache", "fatigue", "cough", "sore throat"]

    symptom_vector = [1 if s in cleaned_data.symptoms else 0 for s in symptom_vocab]

    # Step 3: Preprocess and Validate
    preprocessed_data = PreprocessedPatientData(
        patient_id=cleaned_data.patient_id,
        age=float(cleaned_data.age),
        gender_encoded=gender_map[cleaned_data.gender],
        symptom_vector=symptom_vector
    )

    print("✅ Preprocessed Data Ready:")
    print(preprocessed_data)

except ValidationError as e:
    print("❌ Validation Error:")
    print(e)
```
